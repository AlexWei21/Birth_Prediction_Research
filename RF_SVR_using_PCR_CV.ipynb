```python
import csv
import numpy

with open('Top_200_PCA.csv') as tsvfile:
    reader = csv.reader(tsvfile, dialect = 'excel')
    data_list = []
    for row in reader:
        data_list.append(row)
        
data_list.pop(0)

print (data_list)
```

```python
# Matching Data Member
with open('anoSC1_v11_nokey.csv') as csvfile:
    reader = csv.reader(csvfile)
    GA_Table = []
    for row in reader:
        GA_Table.append(row)

    print(GA_Table[1])

#Could be better
def search(name, where):

    
    for row in range(len(where)):
        if name == where[row][0] :
            return row
    
    return -1
    
for i in range(len(data_list)):
    location = -1
    name_in_HTA20 = data_list[i][0]
    #print (name_in_HTA20)
    location = search(name_in_HTA20, GA_Table)
    #print(location)
    #if location == -1:
        #print ("Error in search")
    data_list[i].append(GA_Table[location][1])

print(data_list [10])
print(len(data_list[10]))

with open('Top_200_PCA_with_Y.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data_list)
```

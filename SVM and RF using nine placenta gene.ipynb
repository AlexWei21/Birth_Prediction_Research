```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
import sklearn
from sklearn import metrics
from sklearn.model_selection import cross_val_score

X_train = df_train.iloc[0:, 1:-1]
Y_train = df_train.iloc[0:, -1]

df = pd.DataFrame(X_train)
df2 = pd.DataFrame(Y_train)

regressor = RandomForestRegressor(n_estimators = 300, random_state=0, n_jobs = -1)

scores = cross_val_score(regressor,X_train,Y_train,cv=10, scoring='neg_mean_squared_error')

print (scores)
```

```python
RMSE = np.sqrt(-1 * scores)
print (RMSE)

sum = 0

for i in range (len(RMSE)):
    sum = sum + RMSE[i]

avg_RMSE = sum / (len(RMSE))

print (avg_RMSE)
```

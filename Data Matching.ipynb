```python
import csv

# Transpose the HTA20_RMA Data
with open('HTA20_RMA.tsv') as tsvfile:
    reader = csv.reader(tsvfile, dialect = 'excel')
    data_list = []

    for row in reader:
        data_list.append(row)
    
    trans = [[data_list[j][i] for j in range(len(data_list))] for i in range(len(data_list[0]))]
        
    print(trans[1])
    print(len(trans[1]))
```

```python
# Matching Data Member
with open('anoSC1_v11_nokey.csv') as csvfile:
    reader = csv.reader(csvfile)
    GA_Table = []
    for row in reader:
        GA_Table.append(row)

    print(GA_Table[1])

#Could be better
def search(name, where):

    
    for row in range(len(where)):
        if name == where[row][0] :
            return row
    
    return -1
    
for i in range(len(trans)):
    location = -1
    name_in_HTA20 = trans[i][0]
    location = search(name_in_HTA20, GA_Table)
    if location == -1:
        print ("Error in search")
    trans[i].append(GA_Table[location][1])

# print(trans [10])
# print(len(trans[10]))

    
```

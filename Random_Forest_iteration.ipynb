```
import csv

# Transpose the HTA20_RMA Data
with open('HTA20_RMA.tsv') as tsvfile:
    reader = csv.reader(tsvfile, dialect = 'excel')
    data_list = []
    for row in reader:
        data_list.append(row)
    
    trans = [[data_list[j][i] for j in range(len(data_list))] for i in range(len(data_list[0]))]
        
    print(trans[1])
    print(len(trans[1]))
```

```
# Matching Data Member
with open('anoSC1_v11_nokey.csv') as csvfile:
    reader = csv.reader(csvfile)
    GA_Table = []
    for row in reader:
        GA_Table.append(row)

    print(GA_Table[1])

#Could be better
def search(name, where):

    
    for row in range(len(where)):
        if name == where[row][0] :
            return row
    
    return -1
    
for i in range(len(trans)):
    location = -1
    name_in_HTA20 = trans[i][0]
    location = search(name_in_HTA20, GA_Table)
    if location == -1:
        print ("Error in search")
    trans[i].append(GA_Table[location][1])

print(trans [10])
print(len(trans[10]))
```

```
# Random Forest Cross Validation with 10 fold

from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
import sklearn
from sklearn import metrics
from sklearn.model_selection import cross_val_score

X_train = df1.iloc[0:, 1:-1]
Y_train = df1.iloc[0:, -1]

df = pd.DataFrame(X_train)
df2 = pd.DataFrame(Y_train)

regressor = RandomForestRegressor(n_estimators = 300, random_state=0, n_jobs = -1)

scores = cross_val_score(regressor,X_train,Y_train,cv=10, scoring='neg_mean_squared_error')

print (scores)
```

```
RMSE = np.sqrt(-1 * scores)
print (RMSE)

sum = 0

for i in range (len(RMSE)):
    sum = sum + RMSE[i]

avg_RMSE = sum / (len(RMSE))

print (avg_RMSE)
```

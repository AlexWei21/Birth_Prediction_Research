```python
import pandas as pd
import csv

with open('Whole_Data_Set_with_Y.csv') as csvfile:
    reader = csv.reader(csvfile, dialect = 'excel')
    data_list = []
    for row in reader:
        data_list.append(row)

print (pd.DataFrame(data_list))
```

```python
data = pd.DataFrame(data_list)
onlydata = data.iloc[0:,1:-1]
print (onlydata)
```

```python
from sklearn.feature_selection import VarianceThreshold

sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
feature_after_selection = sel.fit_transform(onlydata)

print (pd.DataFrame(feature_after_selection))
```

```python
# Random Forest Cross Validation with 10 fold

from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
import sklearn
from sklearn import metrics
from sklearn.model_selection import cross_val_score

X_train = pd.DataFrame(feature_after_selection).iloc[0:,0:]
Y_train = data.iloc[0:,-1]

df = pd.DataFrame(X_train)
df2 = pd.DataFrame(Y_train)

regressor = RandomForestRegressor(n_estimators = 300, random_state=0, n_jobs = -1)

scores = cross_val_score(regressor,X_train,Y_train,cv=10, scoring='neg_mean_squared_error')

print (scores)
```

```python
RMSE = np.sqrt(-1 * scores)
print (RMSE)

sum = 0

for i in range (len(RMSE)):
    sum = sum + RMSE[i]

avg_RMSE = sum / (len(RMSE))

print (avg_RMSE)
```

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVR 

X_train = pd.DataFrame(feature_after_selection).iloc[0:,0:]
Y_train = data.iloc[0:,-1]

df = pd.DataFrame(X_train)
df2 = pd.DataFrame(Y_train)

svregressor = SVR(kernel='rbf', gamma='auto', C=100)

# print (sorted(sklearn.metrics.SCORERS.keys()))

svr_scores = cross_val_score(svregressor,X_train,Y_train,cv=10, scoring='neg_mean_squared_error')

print (svr_scores)
```
